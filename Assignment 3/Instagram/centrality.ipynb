# Centrality Measures

```python
import csv
import networkx as nx
from operator import itemgetter
```

```python
with open('hahanoyume-graph-fixed.csv', 'r') as nodecsv:                      
    nodereader = csv.reader(nodecsv) 
    nodes = [n for n in nodereader][1:]  

degree = [float(n[4]) for n in nodes]
node_names = [n[0] for n in nodes]
closeness = [float(n[11]) for n in nodes]
betweenness = [float(n[13]) for n in nodes]
eigenvector = [float(n[15]) for n in nodes]
```

```python
degree_dict = dict(zip(node_names, degree))
close_dict = dict(zip(node_names, closeness))
between_dict = dict(zip(node_names, betweenness))
eigen_dict = dict(zip(node_names, eigenvector))
```

### Degree Centrality

```python
sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True)
for d in sorted_degree[:20]:
    print(d)
```

```python
import matplotlib.pyplot as plt
import numpy as np

instagramer = list(zip(*sorted_degree[:10]))[0]
degree = list(zip(*sorted_degree[:10]))[1]
x_pos = np.arange(len(instagramer)) 

plt.barh(x_pos, degree, align='center')
plt.yticks(x_pos, instagramer) 
plt.xlabel('Number of Followers')
plt.title('In-degree Centrality')
plt.tight_layout()
plt.savefig('indegree.png')
plt.show()
```

### Eigenvector Centrality

```python
sorted_eigen = sorted(eigen_dict.items(), key=itemgetter(1), reverse=True)
for d in sorted_eigen[:20]:
    print(d)
```

```python
import matplotlib.pyplot as plt
import numpy as np

instagrammer = list(zip(*sorted_eigen[:10]))[0]
eigen = list(zip(*sorted_eigen[:10]))[1]
x_pos = np.arange(len(instagrammer)) 

plt.barh(x_pos, eigen, align='center')
plt.yticks(x_pos, instagrammer) 
plt.xlabel('Eigenvector Score')
plt.title('Eigenvector Centrality')
plt.tight_layout()
plt.savefig('eigen.png')
plt.show()
```

### Betweenness Centrality

```python
sorted_between = sorted(between_dict.items(), key=itemgetter(1), reverse=True)
del sorted_between[:1] # to remove hahanoyume from the centrality analysis
for d in sorted_between[:20]:
    print(d)
```

```python
instagrammer = list(zip(*sorted_between[:10]))[0]
between = list(zip(*sorted_between[:10]))[1]
x_pos = np.arange(len(instagrammer)) 

plt.barh(x_pos, between, align='center')
plt.yticks(x_pos, instagrammer) 
plt.xlabel('Betweenness centrality')
plt.title('Betweenness score')
plt.savefig('betweenness.png')
plt.show()
```

### Closeness Centrality

```python
sorted_close = sorted(close_dict.items(), key=itemgetter(1), reverse=True)
del sorted_close[:4] # to remove hahanoyume from the centrality analysis
for d in sorted_close[:20]:
    print(d)
```

```python
instagrammer = list(zip(*sorted_close[:10]))[0]
close = list(zip(*sorted_close[:10]))[1]
x_pos = np.arange(len(instagrammer)) 

plt.barh(x_pos, close, align='center')
plt.yticks(x_pos, instagrammer) 
plt.xlabel('Closeness score')
plt.title('Closeness centrality')
plt.tight_layout()
plt.savefig('Closeness.png')
plt.show()
```
